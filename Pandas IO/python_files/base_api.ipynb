```python
import pandas as pd
import requests
import json
```

```python
dados_usuarios = requests.get('https://jsonplaceholder.typicode.com/users')
```

```python
resultado = json.loads(dados_usuarios.text)
resultado
```

```python
df_json = pd.DataFrame(resultado)
df_json
```

```python
pd.json_normalize(df_json['address'])
```

```python
pd.json_normalize(df_json['company'])
```
