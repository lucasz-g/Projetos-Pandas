```python
import pandas as pd
```

```python
dados = pd.read_excel('./dados/POP2022_Municipios.xls', header=1, skipfooter=34)
dados
```

```python
dados.groupby('UF').sum(numeric_only=True)
```

```python
dados[dados['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
dados = dados.assign(
    populacao_sem_parenteses=dados['POPULAÇÃO'].replace(r'\(\d{1,2}\)', '', regex=True),
    populacao=lambda x: x['populacao_sem_parenteses'].replace(r'\.', '', regex=True)
)
```

```python
dados[dados['POPULAÇÃO'].str.contains('\(', na=False)]
```

```python
dados = dados.astype({'populacao': 'int64'})
```

```python
dados = dados.groupby('UF')[['populacao']].sum().reset_index()
dados
```
